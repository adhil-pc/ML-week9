```python
def palindrome(n):
    rev=0
    original=n

    while n>0:
        digit=n%10
        rev=rev*10+digit
        n=n//10
    return original==rev
print(f"The number '121' is palindrome :{palindrome(121)}")
print(f"The number '123' is palindrome: {palindrome(123)}")
```

```python
class Shapes:
    def __init__(self,color,is_filled):
        self.color=color
        self.is_filled=is_filled
    def describe(self):
        print(f"This shape is {self.color} color")
class Circle(Shapes):
    def __init__(self,color,is_filled,radius):
        super().__init__(color,is_filled)
        self.radius=radius
    def describe(self):
        super().describe()
        print(f"Area of the circle is {3.14*self.radius*self.radius} cms.")
class Square(Shapes):
    def __init__(self,color,is_filled,width):
        super().__init__(color,is_filled)
        self.width=width
    def describe(self):
        super().describe()
        print(f"The area of square is {self.width*self.width}  cms")

class Triangle(Shapes):
    def __init__(self,color,is_filled,height,width):
        super().__init__(color,is_filled)
        self.height=height
        self.width=width
    def describe(self):
        super().describe()
        print(f"The area of Triangle is {0.5*self.height*self.width} cms")

c=Circle("Red",True,10)
s=Square("Blue",False,10)
t=Triangle(color="Green",is_filled=True,height=20,width=25)

print(s.color)
print(s.is_filled)
print(s.width)

s.describe()
c.describe()
t.describe()
```

```python
from abc import ABC, abstractmethod
class Shapes(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shapes):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return f"Area is:{3.14*self.radius**2}"

class Square(Shapes):
    def __init__(self,side):
        self.side=side
    def area(self):
        return f"Area is :{self.side*self.side}"

class Triangle(Shapes):
    def __init__(self,base,height):
        self.base=base
        self.height=height
    def area(self):
        return f"Area is:{0.5*self.base*self.height}"
class Pizza(Circle):
    def __init__(self,recipe,radius):
        super().__init__(radius)
        self.recipe=recipe
shape=[Circle(5),Square(10),Triangle(10,15),Pizza("Salt",15)]
for i in shape:
    print(f"{i.area()} cms.")
```

```python
class Rectangle:
    def __init__(self,width,height):
        self._height=height
        self._width=width

    @property
    def width(self):
        return f"width is: {self._width:.1f}cm"
    
    @property
    def height(self):
        return f"Height is: {self._height:.1f}cm"
    
    @width.setter
    def width(self,new_width):
        if new_width>0:
            self._width=new_width
        else:
            print("Width must be greater than zero")

    @height.setter
    def height(self,new_height):
        if new_height>0:
            self._height=new_height
        else:
            print("Height must be greater than zero")

r=Rectangle(4,5)
r.width=5
print(r.width)
r.height=10
print(r.height)
r.new_width=10
print(r.new_width)
r.new_height=5
print(r.new_height)
```

```python
class Animal:
    def __init__(self,name):
        self.name=name


class Prey(Animal):
    def flee(self):
        print(f"This {self.name} is fleeing")

class Predator(Animal):
    def hunt(self):
        print(f"This {self.name} is Hunting")

class Rabbit(Prey):
    pass

class Deer(Prey):
    pass

class Tiger(Predator,Prey):
    def color(self):
        print(f"The {self.name} is yellow in color")

r=Rabbit("Momo")
d=Deer("Zoomba")
t=Tiger("Raina")

t.hunt()
t.flee()
t.color()
```

```python
from abc import ABC, abstractmethod
class Sample:
    def __init__(self,name):
        self.name=name

class Vehicle(ABC,Sample):
    @abstractmethod
    def start(self):
        print(f"You Start {self.name}")
    @abstractmethod
    def stop(self):
        print(f"You stop {self.name}")
class Car(Vehicle,Sample):
    def start(self):
        print(f"You Start {self.name}")
    def stop(self):
        print(f"You stop {self.name}")
class Bike(Vehicle,Sample):
    def start(self):
        print(f"You Start {self.name}")
    def stop(self):
        print(f"You stop {self.name}")
    def fuel(self):
        print("The fuel is going to finish")
c=Car("Audi")
c.start()
c.stop()
b=Bike("Bullet")
b.start()
b.stop()
b.fuel()
```

```python
class Shapes:
    def __init__(self,color,is_filled):
        self.color=color
        self.is_filled=is_filled
class Circle(Shapes):
    def __init__(self,color,is_filled,radius):
        super().__init__(color,is_filled)
        self.radius=radius
    def area(self):
        return f"Area of cirle is: {3.14*self.radius**2}"
class Square(Shapes):
    def __init__(self,color,is_filled,side):
        super().__init__(color,is_filled)
        self.side=side
    def area(self):
        return f"Area of Square is: {self.side*self.side}"
class Triangle(Shapes):
    def __init__(self,color,is_filled,base,height):
        super().__init__(color,is_filled)
        self.base=base
        self.height=height
    def area(self):
        return f"Area of triangle is: {0.5*self.base*self.height}"
class Pizza(Circle):
    def __init__(self,color,is_filled,radius,recipe):
        super().__init__(color,is_filled,radius)
        self.recipe=recipe
    def used(self):
        print(f"The recipe used is:{self.recipe}")
    def area(self):
        return f"The area of pizza is: {3.14*self.radius**2}"
c=Circle("Red",True,5)
s=Square("Green",False,3)
t=Triangle("Yellow",True,5,3)
p=Pizza("Red",True,10,"Salt")
p.used()

print(f"{c.area()} \n{s.area()} \n{t.area()} \n{p.area()}")

```

```python
from abc import ABC, abstractmethod
class Shapes(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shapes):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        print(f"The area of circle is: {3.14*self.radius**2}")

class Square(Shapes):
    def __init__(self,side):
        self.side=side
    def area(self):
        print(f"The area of square is: {self.side*self.side}")

class Triangle(Shapes):
    def __init__(self,base,height):
        self.base=base
        self.height=height
    def area(self):
        print(f"The area of triangle is: {0.5*self.base*self.height}")

class Pizza(Circle):
    def __init__(self,topping,radius):
        super().__init__(radius)
        self.topping=topping
    def area(self):
        return f"The Area of pizza is: {3.14*self.radius**2}"
    

shapes=[Circle(5),Square(6),Triangle(7,8),Pizza("Pepper")]
for shape in shapes:
    print(f"{shape.area()} cms.")
```

```python
class Employee:
    def __init__(self,name,position):
        self.name=name
        self.position=position
    
    def get_info(self):
        return f"{self.name} is {self.position}"
    
    @staticmethod
    def is_valid_position(position):
        valid_position=["Manager","Cashier","Cook","Security"]
        return position in valid_position
print(Employee.is_valid_position("Cook"))
e1=Employee("Adhil","Security")
e1.get_info()
print(Employee.is_valid_position("Rocket Launcher"))
```

```python
class Rectangle:
    def __init__(self,length,breadth):
        self._length=length
        self._breadth=breadth

    @property
    def length(self):
        return f"{self._length:.1f}cm"
    
    @property
    def breadth(self):
        return f"{self._breadth:.1f}cm"
    
    @length.setter
    def length(self,new_length):
        if new_length>0:
            self._length=new_length
        else:
            print("Length must be greater than zero")

    @breadth.setter
    def breadth(self,new_breadth):
        if new_breadth>0:
            self._breadth=new_breadth
        else:
            print("Breadth must be greater than zero")

    
r=Rectangle(5,6)
print(f"Length of the rectangle is:{r.length}")
print(f"Breadth of the rectangle iS:{r.breadth}")
print("\nAfter setting new breadth and length")
r.length=10
r.breadth=20
print()
print(f"New length is:{r.length}")
print(f"New breadth is:{r.breadth}")
```
