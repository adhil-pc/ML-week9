```python
double=[x*2 for x in range(1,11)]
print(double)
triple=[x*3 for x in range(1,11)]
print(triple)

fruits=["Apple","Orange","Banana","Strawberry"]
result=[i.upper() for i in fruits]
print(result)
result2=[i[0] for i in fruits]
print(result2)
```

```python
numbers=[1,-2,3,-4,5,-6]
positive=[i for i in numbers if i>0]
negative=[i for i in numbers if i<0]
even=[i for i in numbers if i%2==0]
odd=[i for i in numbers if i%2!=0]
print(positive)
print(negative)
print(even)
print(odd)
```

```python
fruits=("Banana","Apple","Orange","PineApple","Strawberry")
fruits=tuple(sorted(fruits))
print(fruits)
numbers=(2,3,1,4,6,5,7,0,9,8)
numbers=tuple(sorted(numbers,reverse=True))
print(numbers)
```
