```python
#MERGE()=Combine tables using matching columns (Here matching column is "id")
import pandas as pd
df1=pd.DataFrame({"id":[1,2,3],"Name":["Luffy","Zoro","Sanji"]},index=[1,2,3])
df2=pd.DataFrame({"id":[1,2,3],"Marks":[98,99,100]},index=[1,2,3])
print(df1)
print()
print(df2)
print()
print(pd.merge(df1,df2,on="id"))
```

```python
#Join()=Combiene using index
df3=pd.DataFrame({"name":["A","B","C"]},index=[1,2,3])
df4=pd.DataFrame({"Marks":[10,20,30]},index=[1,2,3])
print(df3)
print()
print(df4)
result=df3.join(df4)
print("\n",result)
```

```python
#Concat()=Stack or Append(MEANS:Placing one DF below another)DataFrames
df5=pd.DataFrame({"A":[1,2]})
df6=pd.DataFrame({"A":[3,4]})
print(df5)
print()
print(df6)
print()
print(pd.concat([df5,df6]))#axis=0{Default}
print()
print(pd.concat([df5,df6],axis=1))
```

```python
#hstack()=Combines array side by side: Think of Letf-->Right
import numpy as np
ar1=np.array([1,2,3,4])
ar2=np.array([5,6,7,8])
print(ar1)
print()
print(ar2)
print()
print(np.hstack((ar1,ar2)))

#hsatck()-->HORIZONTAL stack:add columns horiontally

#2D example
ar3=np.array([[1],[2]])
ar4=np.array([[3],[4]])
print()
print(np.hstack((ar3,ar4)))
```

```python
#Vstack()=Combines array from TOP to BOTTOM
ar5=np.array([1,2])
ar6=np.array([3,4])
print(np.vstack((ar5,ar6)))

#vstack--->VERTICAL stack
#2D example
ar7=np.array([[5],[6]])
ar8=np.array([[7],[8]])
print()
print(np.vstack((ar7,ar8)))
```

```python
#shallowcopy()---->creates new outer object but nested objects are still shared
#A new outer list is created but inner list will be the shared Hence any changes made to a/b affects both
import copy
a=[[1,2],[3,4]]
b=copy.copy(a)
b[0][0]=100
print(a)
print()
print(b)

#Deepcopy()--->Copies outer object and also copy nested objects recursively
# A new  outer and inner list is created hence any changes made to c/d did not affect each other
c=[[5,6],[7,8]]
d=copy.deepcopy(c)
d[0][0]=200
print()
print(c)
print()
print(d)
```

```python
#at {similar to loc}-->select data using labels/names, single row/columns possible
import pandas as pd
df=pd.DataFrame({"Name":["A","B","C"],"Marks":[10,20,30]},index=["a","b","c"])
print(df)
print()
print(df.at['a','Marks'])

#iat {Similar to iloc}--->select data using position(Integer based indexing), single row/column at a time
print()
print(df.iat[2,1])
```

```python
#What is a pivot table in Pandas?-->used for rearrange and summarize the dataframe we created
import pandas as pd
df=pd.DataFrame({"Name":["Luffy","Zoro","Zoro","Luffy"],"Subjects":["Science","Maths","Science","Maths"],"Marks":[85,90,95,100]})
print(df)
pivot=pd.pivot_table(df,index="Name",values="Marks",columns="Subjects",aggfunc="mean")
print()
print(pivot)
```

```python
#What are the types of joins (inner, outer, etc.) in Pandas?-->in pandas join are used to combine dataframe based on columns or keys.They used with [ pd.merge()]
#Main types of joins are:
#1)inner joins :keeps only matching values from both data frame
import pandas as pd
df1=pd.DataFrame({"ID": [1,2,3],
                "Name": ["Luffy","Zoro","Sanji"]})
df2=pd.DataFrame({ "ID": [2,3,4],
                "Marks": [90,85,95]})
# print(df1)
# print()
# print(df2)
# print()
result1=pd.merge(df1,df2,on="ID",how="inner")
print(result1)

print()
#2)left joins : keeps all row from left dataframe
result2=pd.merge(df1,df2,on="ID",how="left")
print(result2)

#3)Right joins: keeps all row from the right datarame
print()
result3=pd.merge(df1,df2,on="ID",how="right")
print(result3)

#4) Outer join: Keeps all row from both dataframe.
print()
result4=pd.merge(df1,df2,on="ID",how="outer")
print(result4)
```

```python
#What is the use of grouping in Pandas?-->Devide data into groups and perform operations on each group
import pandas as pd
df=pd.DataFrame({"Name":["Luffy","Zoro","Luffy","Zoro"],"Marks":[80,90,70,100]})
print(df)
print()
group=df.groupby("Name").mean()
print(group)
```

```python
#How are missing values handled? - isna().sum() and fillna()
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Luffy", "Zoro", np.nan],
    "Marks": [90, np.nan, 80]
})
print(df)
print()
print(df.isna())
print()
print(df.isna().sum())
print()
print(df.fillna(0))
print()
result1=df["Marks"]=df["Marks"].fillna(df["Marks"].mean())
result2=df["Name"]=df["Name"].fillna("Sanji")
print(df)
```

```python
#How are duplicate values handled? - drop_duplicates()
import pandas as pd

df = pd.DataFrame({
    "Name": ["Luffy","Zoro","Luffy","Sanji"],
    "Marks": [90,85,90,95]
})
print(df)
result=df.drop_duplicates()
print()
print(result)
```
