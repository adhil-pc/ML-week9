```python
#Python object oriented programming

class Car:
    def __init__(self,model,yom,color,for_sale):
        self.model=model
        self.yom=yom
        self.color=color
        self.for_sale=for_sale
    def start(self):
        print(f"You start the {self.model}")
    def stop(self):
        print(f"You stop the {self.model}")

c1=Car("BMW",2021,"Black",True)
c2=Car("Mercedes",2022,"White",False)
print(c1.model)
print(c2.model)
c1.start()
c2.start()
```

```python
#INHERITANCE
class Animal:
    def __init__(self,name):
        self.name=name
        self.is_alive=True
    def sleep(self):
        print(f"{self.name} is sleeping")
    def eat(self):
        print(f"{self.name} is eating")
    
class Dog(Animal):
    def sounds(self):
        print("Bow")

class Cat(Animal):
    def sounds(self):
        print("Meow")

d=Dog("Scooby")
c=Cat("Poocha")
print(d.name)
print(c.name)
d.sleep()
```

```python
#Multiple and Multilevel inheritance
class Animal:
    def __init__(self,is_alive):
        self.is_alive=is_alive

class Predator(Animal):
    def hunt(self):
        return "Animal is hunting"

class Prey(Animal):
    def flee(self):
        return "Animal is fleeing"

class Cat(Prey):
    pass

class Tiger(Predator):
    pass

class Fish(Prey,Predator):
    pass


c=Cat(True)
t=Tiger(False)
f=Fish(True)

f.flee()
f.hunt()
```

```python
#Abstract Class
from abc import ABC, abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def go(self):
        return "The car is going"
    
    @abstractmethod
    def stop(self):
        return "The car is stopped"

class Car(Vehicle):
    def go(self):
        return "The car is going"
    def stop(self):
        return "The car is stopped"

c=Car()
c.go()
```

```python
class Shapes:
    def __init__(self,is_filled):
        self.is_filled=is_filled

class Circle(Shapes):
    def __init__(self,radius,is_filled):
        super().__init__(is_filled)
        self.radius=radius
    def info(self):
        return f"Radius is {self.radius} and it is filled {self.is_filled}"
    
    def area(self):
        return f"Area of circle is {3.14*self.radius**2}"

class Square(Shapes):
    def __init__(self,side,is_filled):
        super().__init__(is_filled)
        self.side=side
    def info(self):
        return f"Side is {self.side} and is filled {self.is_filled}"
    def area(self):
        return f"The area of square is {self.side*self.side}"

class Triangle(Shapes):
    def __init__(self,base,height,is_filled):
        super().__init__(is_filled)
        self.base=base
        self.height=height
    def info(self):
        return f"The Height and base of triangle is {self.height} and {self.base} and is filled {self.is_filled}"
    def area(self):
        return f"The area of triangle is {0.5*self.base*self.height}"
c=Circle(5,True)
s=Square(10,False)
t=Triangle(15,20,False)
c.info()
c.area()
```

```python
#Static method
class Employee:
    def __init__(self,name,position):
        self.name=name
        self.position=position
    def info(self):
        print(f'{self.name} is {self.position}')
    @staticmethod
    def is_valid_position(position):
        valid_position=["Manager","Cook","Security"]
        return position in valid_position
    
e1=Employee("Luffy","Pirate King")
e2=Employee("Sanji","Cook")
e2.info()
print(Employee.is_valid_position("Cook"))
```

```python
#Class Method
class Student:
    count=0
    def __init__(self,name,gpa):
        self.name=name
        self.gpa=gpa
        Student.count+=1
    def info(self):
        return f"{self.name} have {self.gpa} gpa"
    
    @classmethod
    def get_count(cls):
        return f"There are {cls.count} in this class"
    
s1=Student("Luffy",5)
s2=Student("zoro",4)
s3=Student("Sanji",3)
print(Student.get_count())
```
