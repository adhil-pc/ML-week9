```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

product_ids=np.random.randint(1001,1011,100)
quantities=np.random.randint(1,5,100)
prices=np.round(np.random.uniform(10,100,100),2)
set_date=pd.date_range(start='2025-1-1',end='2025-1-10')
dates=np.random.choice(set_date,100)

missing_value=np.random.choice(len(prices),5)
prices[missing_value]=np.nan

sales_df=pd.DataFrame({'ProductID':product_ids,'Quantity':quantities,'Price':prices,'Date':dates})

sales_df['TotalSale']=sales_df['Quantity']*sales_df['Price']



print(sales_df.head(10))
print()
print(sales_df['Price'].isna().sum())
print()
sales_df['Price']=sales_df['Price'].fillna(sales_df['Price'].mean())
print(sales_df['Price'].isna().sum())


group=sales_df.groupby("ProductID").agg({'Quantity':'sum','Price':'sum','TotalSale':'sum'})
print(group)
print()

figure,axes=plt.subplots(1,2)
axes[0].bar(group.index,group['Price'])
axes[0].set_title('Total revenue per ProductID')
axes[0].set_xlabel('Product ID')
axes[0].set_ylabel('Revenue')

axes[1].hist(group['TotalSale'],bins=10,color='Green',edgecolor='Black')
axes[1].set_title('Total revenue per ProductI')
axes[1].set_xlabel('Total Sale')
axes[1].set_ylabel('Product ID')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

Employee_ID=np.random.randint(1,501,500)
Department=np.random.choice(["HR",'IT','Sales',"Finance"],500)
Salary=np.round(np.random.randint(10000,500001,500),2).astype(float)
Experience_years=np.random.uniform(1,21,500)
Performance_Score=np.random.randint(1,6,500)

missing_value=np.random.choice(len(Salary),int(len(Salary)*0.10),replace=False)
Salary[missing_value]=np.nan

missing_expyr=np.random.choice(len(Experience_years),int(len(Experience_years)*0.10),replace=False)
Experience_years[missing_expyr]=np.nan

report=pd.DataFrame({"ID":Employee_ID,"DEPT":Department,"SALARY":Salary,"EXPERIENCE":Experience_years,"PERFORMANCE":Performance_Score})
print(report.head(10))

print()
print("Total missing values in SALARY: ",report['SALARY'].isna().sum())
print("Total missing values in EXPERIENCE: ",report['EXPERIENCE'].isna().sum())

print()
print("AFTER FILLING MISSING VALUES")
print()

report["SALARY"]=report.groupby("DEPT")["SALARY"].transform(lambda x:x.fillna(x.mean()))
report["EXPERIENCE"]=report["EXPERIENCE"].fillna(report["EXPERIENCE"].mean())
print("Total missing values in SALARY: ",report['SALARY'].isna().sum())
print("Total missing values in EXPERIENCE: ",report['EXPERIENCE'].isna().sum())

print()
group=report.groupby("DEPT")[['SALARY','EXPERIENCE','PERFORMANCE']].agg({"SALARY":'mean',"EXPERIENCE":"mean","PERFORMANCE":'mean'})
print(group)

print()
plt.figure()
plt.bar(group.index,group['SALARY'])
plt.title("Salary per department")
plt.xlabel("DEPARTMENT")
plt.ylabel("SALARY")

plt.figure()
report.boxplot(column='SALARY',by='DEPT',color='Red')

plt.figure()
plt.scatter(report["EXPERIENCE"],report["SALARY"],c=report["PERFORMANCE"])
plt.title("EXPERIENCE YR VS SALARY")
plt.xlabel("EXP")
plt.ylabel("SALARY")
plt.tight_layout()
plt.show()
```
