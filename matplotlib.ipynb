```python
import matplotlib.pyplot as plt #importing library
import numpy as np
x=np.array([2023,2024,2025,2026])
y1=np.array([20,31,40,50])
y2=np.array([11,32,45,76])
y3=np.array([90,78,99,22])

line_style=dict(marker="o",
               markersize=10,
               markerfacecolor="Yellow",
               markeredgecolor='Black',
               linestyle="dotted",
               linewidth=2)#crate dictionary(not mandatory) to set the line style

plt.title("Class Size",fontsize=20,color="orange")#to set the heading of plot

plt.xlabel("<--Year-->",fontsize=15,
                   family="Arial",
                    fontweight="bold",
                    color="Black") #To set the label of x axis
plt.ylabel("<--Strenght of the class-->",fontsize=15,
           fontweight="bold",
           color="Black",
           family="Arial")#To set the label of y axis

plt.tick_params(axis="both",colors="black")#To set the parameters that er given to x and y axes

plt.xticks(x)#To remove the extra entities other than what we enter

plt.plot(x,y1,color="Green",**line_style)#.plot() used to plot the entire graph
plt.plot(x,y2,color="Blue",**line_style)
plt.plot(x,y3,color="red",**line_style)
plt.show()#TO DISPLAY  THE GRAPH
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([10,20,30,40,50])
y=np.array([15,25,35,45,55])
plt.grid(axis="both",linewidth=1,
         color='red',
         linestyle="dotted")#Helps plot easier to read since grid is used
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y)
plt.show()
```

```python
#Bar chart represent a data by representing each category using a bar
import matplotlib.pyplot as plt
import numpy as np
categories=np.array(["Grains","Vegetables","Fruits","Protien","Diary","Sweets"])
values=np.array([1,3,2,4,1.5,5])
plt.title("Dialy Consumption")
plt.xlabel("<--Categories-->",fontsize=10,fontweight='bold')
plt.ylabel("<--Values-->",fontsize=10,fontweight='bold')
plt.bar(categories,values,color="yellow")
plt.show()
```

```python
#Pie chart-->Circular chart divided into slices to show percentage of total
import matplotlib.pyplot as plt
import numpy as np
categories=np.array(["1st Years","2nd Years","3rd Years","4th Years"])
value=np.array([200,197,203,211])
color=np.array(["Red","Green","orange","Yellow"])
plt.pie(value,labels=categories,
               autopct='%1.1f%%',
               colors=color,
               explode=[0,0,0,0.1],
               shadow=True,
               startangle=90)
plt.title("XYZ college")
plt.show()
```

```python
#Scatter graph--> They show the relationship between 2 variables
#e.g., Number of hours study >> Test grade
import matplotlib.pyplot as plt
import numpy as np
x1=np.array([0,1,1,2,3,4,5,6,7,7])
y1=np.array([50,60,63,70,72,78,81,88,90,95])

x2=np.array([0,0,1,2,3,3,4,5,6,7])
y2=np.array([51,50,60,65,66,70,73,80,87,90])

plt.scatter(x1,y1,color="Green",alpha=0.5,s=100,label="Class A")#label is essential to use LEGEND()
plt.scatter(x2,y2,color="Red",alpha=0.3,s=80,label="Class B")

plt.title("Test Score",fontweight='bold')
plt.xlabel("<--Hours Studies-->",fontweight='bold')
plt.ylabel("<--Grade-->",fontweight='bold')

plt.legend()
plt.show()
```

```python
#Histogram-->A visual representation of the distribution of quantitative data.
#            They group values in to intervals(bin)
#            and counts how many fall in range.
import matplotlib.pyplot as plt
import numpy as np
scores=np.random.normal(loc=80,scale=10,size=100)
scores=np.clip(scores,0,100)
plt.hist(scores,bins=10,color="lightgreen",edgecolor="black")
plt.title("Exam Scores",fontweight='bold')
plt.xlabel("Scores",fontweight='bold')
plt.ylabel("Number of students",fontweight='bold')
plt.show()
```

```python
#Subplots
import numpy as np
import matplotlib.pyplot as plt
figure,axes=plt.subplots(2,2)
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,7,10])
axes[0,0].plot(x,x*2,color="Red")
axes[0,0].set_title("x*2 Graph")

axes[0,1].plot(x,y,color="Green")
axes[0,1].set_title("Simple Graph")

axes[1,0].plot(y,y**2,color="Brown")
axes[1,0].set_title("Squared Graph")

axes[1,1].plot(x,x**2,color="lightblue")
axes[1,1].set_title("Squared Graph")

plt.tight_layout()
plt.show()
```
