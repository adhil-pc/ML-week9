FINDING ANGLE BETWEEN TWO VECTORS

```python
A=[1,2]
B=[4,6]
#cosx=(AB)/(|A||B|)
dot=A[0]*B[0]+A[1]*B[1]
magA=((A[0]**2)+(A[1]**2))**0.5
magB=((B[0]**2)+(B[1]**2))**0.5
cosx=(dot/(magA*magB))
print(f'Cos theta is:{cosx}')
import math
theta=math.degrees(math.acos(cosx))
print(f'Angle bw 2 vectors is:{theta}')
```

FINDING MINOR, COFACTOR AND INVERSE(IMPORTANT) OF A MATRIX

```python
#MINOR OF ELEMENT A[2][2]
A = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
minora33=[]
row=2
col=2
for i in range(len(A)):
    if i==row:
        continue
    new_row=[]
    for j in range(len(A[0])):
        if j==col:
            continue
        new_row.append(A[i][j])
    minora33.append(new_row)
print(F'Minor Matrix:{minora33}')

minor_9=[A[0][0]*A[1][1]-A[0][1]*A[1][0]]
print(f'Minor of 9 is:{minor_9}')

#COFACTOR
sign=(-1)**(row+col)
cofactor=sign*minor_9
print(f'Cofactor is:{cofactor}')


#FINDING INVERSE OF A MATRIX USING GAUSS JORDAN METHOD IS EASY COMPPARED TO TRADITIONAL METHOD
#GJ method
# A =
# [2 1 1]
# [1 3 2]
# [1 0 0]
#
# I =
# [1 0 0]
# [0 1 0]
# [0 0 1]
#
# Augmented matrix [A | I]
print()
A=[[2,1,1,1,0,0],
   [1,3,2,0,1,0],
   [1,0,0,0,0,1]]
A[2],A[0]=A[0],A[2]

for i in range(len(A[0])):
    A[1][i]=A[1][i]-A[0][i]

for i in range(len(A[0])):
    A[2][i]=A[2][i]-2*A[0][i]

for i in range(len(A[0])):
    A[1][i]=A[1][i]-2*A[2][i]

for i in range(len(A[0])):
    A[2][i]=A[2][i]-A[1][i]

for row in A:
    print(f'After rearranging :{row}')
print()
#[1, 0, 0, 0, 0, 1]
#[0, 1, 0, -2, 1, 3]
#[0, 0, 1, 3, -1, -5]
#This is the result we got and the matrix other than identity matrix is consider as the inverse of matrix A
inverse=[[A[0][3],A[0][4],A[0][5]],
         [A[1][3],A[1][4],A[1][5]],
         [A[2][3],A[2][4],A[2][5]]]
print(f'Final result is:{inverse}')
```

FINDING COVARIANCE OF 2*2 MATRIX

```python
height=[170,175,180]
weight=[65,70,75]

mean_height=sum(height)/len(height)
mean_weight=sum(weight)/len(weight)

var_height=0
var_weight=0
cov=0

for i in range(len(height)):
    var_height+=(height[i]-mean_height)**2
    var_weight+=(weight[i]-mean_weight)**2
    cov+=((height[i]-mean_height)*(weight[i]-mean_weight))
var_height=var_height/(len(height)-1)
var_weight=var_weight/(len(height)-1)
cov=cov/(len(height)-1)

cov_matrix=[[cov,var_height],
            [var_weight,cov]]
print(f'Covariance matrix is:{cov_matrix}')
```

PCA(PRINCIPAL COMPONENT ANALYSIS)--->(IMPORTANT)

```python
import numpy as np

X = np.array([
    [90, 88, 91],
    [80, 82, 79],
    [70, 68, 72],
    [60, 62, 61]
])

#s1:MEAN
mean=np.mean(X,axis=0)
print(f'Mean is:{mean}')

#s2:CENTER DATA
X_centered=X-mean
print()
print(f'Centered Data is:{X_centered}')

#s3:COVARIANCE MATRIX
print()
cov_matrix=np.cov(X_centered.T)
print(f'Covarience Matrix:{cov_matrix}')

#s4:EIGEN VALUES?VECTORS
Eigenvalues,Eigenvectors=np.linalg.eig(cov_matrix)
print()
print(f'Eigen Values:{Eigenvalues}',
      f'\nEigen Vectors:{Eigenvectors}')

#s5:Sorting them
idx=np.argsort(Eigenvalues)[::-1]
Eigenvalues=Eigenvalues[idx]
Eigenvectors=Eigenvectors[:,idx]
print()
print(f'Sorted Eigen VEctors are:{Eigenvectors}')
print()

#s6:Display principal Component
w=Eigenvectors[:,0]
print(f'Principal Component:{w}')

#s7:Print final result
print()
z=X_centered*w
print(f'PCA result is:{z}')
```

GRADIENT DESCEND -->(IMPORTANT)

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5])
y=np.array([20,40,60,80,100])
#y=wx+b
w=0
b=0
learning_rate=0.01
epochs=1000

#Learning model
for epoch in range(epochs):
    #implementing prediction
    y_pred=w*x+b
    error=y_pred-y
    #fiding error
    cost=np.mean(error**2)

    dw=(2/len(x))*np.sum(error*x)
    db=(2/len(y))*np.sum(error)

    w=w-learning_rate*dw
    b=b-learning_rate*db

print(f'Weight:{w}')
print(f'Bias:{b}')
print(f'Cost:{cost}')

plt.figure()
plt.scatter(x,y)
plt.plot(x,w*x+b)
plt.title("REPORT")
plt.xlabel("Study hours")
plt.ylabel("Marks")
plt.show()
```
