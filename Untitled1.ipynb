```python
def palindrome(n):
    rev=0
    original=n

    while n>0:
        digit=n%10
        rev=rev*10+digit
        n=n//10
    return original==rev
print(f"The number '121' is palindrome :{palindrome(121)}")
print(f"The number '123' is palindrome: {palindrome(123)}")
```

```python
from abc import ABC, abstractmethod
class Calculator(ABC):
    @abstractmethod
    def add(self,a,b):
        pass
    @abstractmethod
    def sub(self,a,b):
        pass
    @abstractmethod
    def mul(self,a,b):
        pass
    @abstractmethod
    def div(self,a,b):
        pass
class Operation(Calculator):
    def add(self,a,b):
        return a+b
    def sub(self,a,b):
        return a-b
    def mul(self,a,b):
        return a*b
    def div(self,a,b):
        if b==0:
            print("ZeroDivisionError")
        return a/b
obj=Operation()
print("Addition is:",obj.add(10,5))
print("Subtraction is:",obj.sub(10,5))
print("Multipication is:",obj.mul(10,5))
print("Division is:",obj.div(10,5))
```

```python
pantry = {
    "chicken": 500,
    "lemon": 2,
    "cumin": 24,
    "paprika": 18,
    "chilli powder": 7,
    "yogurt": 300,
    "oil": 450,
    "onion": 5,
    "garlic": 9,
    "ginger": 2,
    "tomato puree": 125,
    "almonds": 75,
    "rice": 500,
    "coriander": 20,
    "lime": 3,
    "pepper": 8,
    "egg": 6,
    "pizza": 2,
    "spam": 1,
}

recipes = {
    "Butter chicken": [
        "chicken",
              "lemon",
        "cumin",
        "paprika",
        "chilli powder",
        "yogurt",
        "oil",
        "onion",
        "garlic",
        "ginger",
        "tomato puree",
        "almonds",
        "rice",
        "coriander",
        "lime",
    ],
    "Chicken and chips": [
        "chicken",
        "potatoes",
        "salt",
        "malt vinegar",
    ],
    "Pizza": [
        "pizza",
    ],
    "Egg sandwich": [
        "egg",
        "bread",
        "butter",
    ],
    "Beans on toast": [
        "beans",
        "bread",
    ],
    "Spam a la tin": [
        "spam",
        "tin opener",
        "spoon",
    ],
}
while True:
    print("1.Show Recipes")
    print("2.For Cook Recipes")
    print("3.For Show Pantry")
    print("4. EXIT")
    choice=int(input("Enter Your Choice"))


    if choice==1:
        print("\nAvailable Recipes")
        for recipe in recipes:
            print(recipe)

    elif choice==2:
        recipe_name=input("Enter recipe name")
        if recipe_name in recipes:
            ingredients=recipes[recipe_name]
            can_cook=True

            for item in ingredients:
                if item not in pantry or pantry[item]<=0:
                    print(item,"Is not available")
                    can_cook=False

            if can_cook:
                print("Cooking",recipe_name)
                for item in ingredients:
                    pantry[item]-=1
            print("Recipe Completed successfully")
        else:
            print("Recipe not found")

    elif choice==3:
        print("\nPantry items")
        for item,quantity in pantry.items():
            print(item, ":", quantity)
    elif choice == 4:
        print("Exiting Program")
        break
    else:
        print("Invalid Choice")
```
