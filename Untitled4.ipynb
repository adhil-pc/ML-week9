```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

report=pd.read_csv('sales.csv')
missing_values=report.isna().sum()

print(missing_values)
print()

report['Units_Sold']=report['Units_Sold'].fillna(report['Units_Sold'].mean())
report['Revenue']=report['Revenue'].fillna(report['Revenue'].mean())

missing_values=report.isna().sum()
print(missing_values)
print()

report['Date']=pd.to_datetime(report['Date'])
report['Year']=report['Date'].dt.year
report['Month']=report['Date'].dt.month

report['RPU']=report['Revenue']/report['Units_Sold']

top_products=report.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(3)
print(top_products)
print()

avg_SPM=report.groupby('Region')['Revenue'].mean()
print(avg_SPM)
print()

group=report.groupby('Product').agg({'Revenue':'sum','Units_Sold':'sum'})
print(group)
print()

mean=np.mean(report['Revenue'])
median=np.median(report["Revenue"])
std=np.std(report['Revenue'])
print(f'Mean is:{mean}, Median is: {median}, STD is: {std}')
print()

numpy_array=np.array(report['RPU'])
print(numpy_array)
print()
print(report.head(10))

figure,axes=plt.subplots(1,3)
axes[0].bar(avg_SPM.index,avg_SPM.values)
axes[0].set_title('Total Ravenue per Region')
axes[0].set_xlabel('Region')
axes[0].set_ylabel('Revenue')

monthly_trnd=report.groupby('Month')['Revenue'].sum()
print(monthly_trnd)

axes[1].plot(monthly_trnd.index,monthly_trnd.values,color='Green',linestyle='dotted',marker='o')
axes[1].set_title('Monthly Trend')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Revenue')

colors={
    "North":'Red',
    "South":'Blue',
    'East':'Green',
    'West':'Orange'
}
axes[2].scatter(report['Units_Sold'],report['Revenue'],c=report['Region'].map(colors))
axes[1].set_title("Revenue as per unit sold")
axes[1].set_xlabel("Unit Sold")
axes[1].set_ylabel('Revenue')
plt.tight_layout()
plt.show()
```
