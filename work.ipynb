```python
height=[170,175,180]
weight=[65,70,75]

mean_height=sum(height)/len(height)
mean_weight=sum(weight)/len(weight)

var_height=0
var_weight=0
cov=0

for i in range(len(height)):
    var_height+=(height[i]-mean_height)**2
    var_weight+=(weight[i]-mean_weight)**2
    cov+=((height[i]-mean_height)*(weight[i]-mean_weight))
var_height=var_height/(len(height)-1)
var_weight=var_weight/(len(height)-1)
cov=cov/(len(height)-1)

cov_matrix=[[cov,var_height],
            [var_weight,cov]]
for row in cov_matrix:
    print(cov_matrix)
```
