```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,4,5])
y=np.array([20,40,60,80,100])
#y=wx+b
w=0
b=0
learning_rate=0.01
epochs=1000

#Learning model
for epoch in range(epochs):
    #implementing prediction
    y_pred=w*x+b
    error=y_pred-y
    #fiding error
    cost=np.mean(error**2)

    dw=(2/len(x))*np.sum(error*x)
    db=(2/len(y))*np.sum(error)

    w=w-learning_rate*dw
    b=b-learning_rate*db

print(f'Weight:{w}')
print(f'Bias:{b}')
print(f'Cost:{cost}')

plt.figure()
plt.scatter(x,y)
plt.plot(x,w*x+b)
plt.title("REPORT")
plt.xlabel("Study hours")
plt.ylabel("Marks")
plt.show()
    
```

```python
#PCA
import numpy as np
X=np.array([[90,92,92,88],
           [88,99,97,94],
           [90,94,97,70],
           [85,79,95,94]])
#Fining mean
mean=np.mean(X,axis=0)
print(f'Mean is:{mean}')
print()

#Centering Data
X_centered=X-mean
print(f'Data after Centering:{X_centered}')
print()

#Determining covariance matrix
cov_matrix=np.cov(X_centered.T)
print(f'Covariance matrix is:{cov_matrix}')
print()

#FInding EIGEN Values/vectors
eigenvalues,eigenvectors=np.linalg.eig(cov_matrix)
print(f'EigenValues:{eigenvalues}',
      f'\n',
      f'\nEigenVectors:{eigenvectors}')
print()

#Sorting Eigen value and vectors for easy understanding and for determine the Principal component
sort=np.argsort(eigenvectors)[::-1]
eigenvalues=eigenvalues[sort]
eigenvectors=eigenvectors[:,sort]

#Finding the principal component
w=eigenvectors[:,0]
print(f'Principal component:{w}')
print()

#Fianl result
z=np.dot(w,X_centered)
print(f'Final PCA result is:{z}')
```
