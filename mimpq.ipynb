```python
#MERGE()=Combine tables using matching columns (Here matching column is "id")
import pandas as pd
df1=pd.DataFrame({"id":[1,2,3],"Name":["Luffy","Zoro","Sanji"]},index=[1,2,3])
df2=pd.DataFrame({"id":[1,2,3],"Marks":[98,99,100]},index=[1,2,3])
print(df1)
print()
print(df2)
print()
print(pd.merge(df1,df2,on="id"))
```

```python
#hstack()=Combines array side by side: Think of Letf-->Right
import numpy as np
ar1=np.array([1,2,3,4])
ar2=np.array([5,6,7,8])
print(ar1)
print()
print(ar2)
print()
print(np.hstack((ar1,ar2)))

#hsatck()-->HORIZONTAL stack:add columns horiontally

#2D example
ar3=np.array([[1],[2]])
ar4=np.array([[3],[4]])
print()
print(np.hstack((ar3,ar4)))
```

```python
#shallowcopy()---->creates new outer object but nested objects are still shared
#A new outer list is created but inner list will be the shared Hence any changes made to a/b affects both
import copy
a=[[1,2],[3,4]]
b=copy.copy(a)
b[0][0]=100
print(a)
print()
print(b)

#Deepcopy()--->Copies outer object and also copy nested objects recursively
# A new  outer and inner list is created hence any changes made to c/d did not affect each other
c=[[5,6],[7,8]]
d=copy.deepcopy(c)
d[0][0]=200
print()
print(c)
print()
print(d)
```

```python
#What are the types of joins (inner, outer, etc.) in Pandas?-->in pandas join are used to combine dataframe based on columns or keys.They used with [ pd.merge()]
#Main types of joins are:
#1)inner joins :keeps only matching values from both data frame
import pandas as pd
df1=pd.DataFrame({"ID": [1,2,3],
                "Name": ["Luffy","Zoro","Sanji"]})
df2=pd.DataFrame({ "ID": [2,3,4],
                "Marks": [90,85,95]})
# print(df1)
# print()
# print(df2)
# print()
result1=pd.merge(df1,df2,on="ID",how="inner")
print(result1)

print()
#2)left joins : keeps all row from left dataframe
result2=pd.merge(df1,df2,on="ID",how="left")
print(result2)

#3)Right joins: keeps all row from the right datarame
print()
result3=pd.merge(df1,df2,on="ID",how="right")
print(result3)

#4) Outer join: Keeps all row from both dataframe.
print()
result4=pd.merge(df1,df2,on="ID",how="outer")
print(result4)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

product_id=np.random.randint(1001,1011,100)
quantities=np.random.randint(1,6,100)#Generates random Integer numbers between a range.
prices=np.round(np.random.uniform(10,100,100),2)#Generates random floating-point numbers between a range.
set_date=pd.date_range(start='2025-1-1',end='2025-1-10')
dates=np.random.choice(set_date,100)

#rand:Generates random floating-point numbers between a 0 and 1.

sales_df=pd.DataFrame({"ProductID":product_id,"Quantity":quantities,"Price":prices,"Date":dates})
sales_df['TotalSale']=sales_df["Quantity"]*sales_df["Price"]

missing_value=np.random.choice(len(prices),5)
prices[missing_value]=np.nan

sales_df["Price"]=sales_df["Price"].fillna(sales_df["Price"].mean())

grouping=sales_df.groupby(product_id).agg({"Quantity":"sum","Price":"sum"})

print(grouping)
print()

print(sales_df.head(10))

figure,axes=plt.subplots(1,2)
axes[0].bar(grouping.index,grouping.Quantity,color='skyblue')
axes[0].set_title("Revnue Of The Year")
axes[0].set_xlabel("Index")
axes[0].set_ylabel("Quantity")

axes[1].hist(grouping.Price,color='Green',bins=10,edgecolor='black')
axes[1].set_title("Quantity sold")
axes[1].set_xlabel("Index")
axes[1].set_ylabel("Price")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

employee_id=np.random.randint(1,501,500)
department=np.random.choice(["HR","IT","Finance","Sales"],500)
salary=np.round(np.random.uniform(10000,50000,500),2)
experience_yr=np.random.randint(1,21,500).astype(float)
performance_score=np.random.randint(1,6,500)


missing_value=np.random.choice(len(salary),int(len(salary)*0.10),replace=False)
salary[missing_value]=np.nan

missing_year=np.random.choice(len(experience_yr),int(len(experience_yr)*0.10),replace=False)
experience_yr[missing_year]=np.nan

df=pd.DataFrame({"ID":employee_id,"Department":department,"Salary":salary,"Experience":experience_yr,"Performance":performance_score})

df['Salary']=df.groupby("Department")["Salary"].transform(lambda x:x.fillna(x.median()))
df['Experience']=df["Experience"].fillna(df["Experience"].mean())

group=df.groupby("Department").agg({"Salary":"mean","Performance":"mean","Experience":"mean"})
print(group)

print(df.isna().sum())
print()
plt.figure()
plt.bar(group.index,group.Salary,color="skyblue")
plt.xlabel("Department")
plt.ylabel("Salary")
plt.title("Average Salary Per Department")
print()
plt.figure()
df.boxplot(column='Salary',by="Department")
plt.figure()
plt.scatter(experience_yr,salary,c=performance_score)#If the entities(here:Performance score) are numerical values will give colors automatically :::: 1-->Dark colour::5-->Bright colos

plt.tight_layout()
print(df.head(10))
print()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file=pd.read_csv("sales.csv")

file['Revenue']=file['Revenue'].fillna(file['Revenue'].mean())
file['Units_Sold']=file['Units_Sold'].fillna(file['Units_Sold'].median())

print(file.isna().sum())
print()

file['Date']=pd.to_datetime(file['Date'])
file['Year']=file['Date'].dt.year
file['Month']=file['Date'].dt.month

file['Revenue/sold']=(file['Revenue']/file['Units_Sold'])
print(file.head(10))
print()
top_products=file.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(3)
print(top_products)
print()

monthly_sales=file.groupby(['Region','Month'])['Units_Sold'].mean()
print(monthly_sales)
print()

grouped=file.groupby('Product').agg({"Revenue":"sum","Units_Sold":"mean"})
print(grouped)
print()

mean_revenue=np.mean(file['Revenue'])
print('Mean Revenue is:',mean_revenue)
median_revenue=np.median(file['Revenue'])
print('Median Revenue is:',median_revenue)
std_revenue=np.std(file['Revenue'])
print('STD Revenue is:',std_revenue)
revenue_array=np.array(file['Revenue'])
print("Revenue array is:",revenue_array)
print()

revenue_region=file.groupby("Region")['Revenue'].sum()
print(revenue_region)
print()
plt.figure()
plt.bar(revenue_region.index,revenue_region.values, color='Green')
plt.title("Plot for Revenue Per Region")
plt.xlabel("Revenue")
plt.ylabel("Region")

revenue_month=file.groupby('Month')['Revenue'].sum()
print(revenue_month)
print()

plt.figure()
plt.plot(revenue_month.index,revenue_month.values,color='Red',linestyle='dotted',marker='o',mfc='Yellow')
plt.title("Plot for Revenue Per Month")
plt.xlabel("Revenue")
plt.ylabel("Month")

colors={
    "North":'Red',
    "South":'Blue',
    'East':'Green',
    'West':'Orange'
}
plt.figure()
plt.scatter(file['Units_Sold'],file['Revenue'],c=file['Region'].map(colors))#If the entities(here:Region) are numerical values will give colors automatically :::: 1-->Dark colour::5-->Bright colos
plt.title("Revenue as per unit sold")
plt.xlabel("Unit Sold")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```
